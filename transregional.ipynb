```python
df['contains_islamic_state'] = df['gname'].str.contains('Islamic State', case=False, na=False)
```

```python
df['contains_islamic_state']
```

```python
ISIS_columns = df[df['contains_islamic_state'] == True]
count = len(ISIS_columns)
```

```python
print("Count of rows with 'Islamic State':", count)
```

```python
import folium
from folium import plugins

# Drop rows with NaN values in latitude or longitude
ISIS_columns = ISIS_columns.dropna(subset=['latitude', 'longitude'])

# Create a folium map centered on a location
map_center = [ISIS_columns['latitude'].mean(), ISIS_columns['longitude'].mean()]
map_zoom = 2
map_osm = folium.Map(location=map_center, zoom_start=map_zoom)

# Create a list of latitudes and longitudes
locations = ISIS_columns[['latitude', 'longitude']].values

# Create a heatmap layer based on the locations
heat_map = plugins.HeatMap(locations)
map_osm.add_child(heat_map)

# Display the map
map_osm
```

```python
import folium
from folium import plugins

# Exclude Iraq and Syria locations from the DataFrame
ISIS_columns = ISIS_columns[~ISIS_columns['natlty1_txt'].isin(['Iraq', 'Syria'])]

# Create a folium map centered on a location
map_center = [ISIS_columns['latitude'].mean(), ISIS_columns['longitude'].mean()]
map_zoom = 2
map_osm = folium.Map(location=map_center, zoom_start=map_zoom)

# Create a list of latitudes and longitudes
locations = ISIS_columns[['latitude', 'longitude']].values

# Create a heatmap layer based on the locations
heat_map = plugins.HeatMap(locations)
map_osm.add_child(heat_map)

# Display the map
map_osm
```
